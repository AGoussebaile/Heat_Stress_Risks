```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data points (temperature in Celsius and percentage work loss)
temperature = np.array([-20, 0, 25, 30, 35, 40, 50, 70])
work_loss = np.array([0, 0, 0, 0.05, 0.40, 0.80, 1, 1])

# Function for piecewise linear interpolation
def piecewise_linear_interpolation(x, temperature, work_loss):
    return np.interp(x, temperature, work_loss)

# Test the function and plot the results
x_values = np.linspace(-20, 70, 500)  # Range of temperatures
y_values = piecewise_linear_interpolation(x_values, temperature, work_loss)

# Plotting
plt.plot(x_values, y_values, label='Piecewise Linear Function')
plt.scatter(temperature, work_loss, color='red', label='Data Points')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Percentage Work Loss')
plt.title('Temperature vs. Percentage Work Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def work_loss_function(temp):
    if temp < 25:
        return 0.0
    elif temp > 45:
        return 1.0
    else:
        return (temp - 25.0) / 20.0 

# Vectorize the function to apply it over an array
work_loss_function_vectorized = np.vectorize(work_loss_function)

# Test the function over a range of temperatures
temperature_range = np.linspace(0, 60, 500)
work_loss_values = work_loss_function_vectorized(temperature_range)

# Plotting
plt.plot(temperature_range, work_loss_values, label='Work Loss Function')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Percentage Work Loss')
plt.title('Temperature vs. Percentage Work Loss')
plt.legend()
plt.grid(True)
plt.show()
```
